```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
age = [21,20,22,22,23,21,25,30,31]
```

```python
income = [60,55,60,61,65,62,65,70,68]
```

```python
dict = {
    'age' : age,
    'income' : income
}
```

```python
data = pd.DataFrame(dict)
```

```python
data
```

```python
x = data[['age']]
x
```

```python
y = data.income
y
```

```python
knn.fit(x,y)
```

```python
knn_predict=knn.predict([[22]])
```

```python
knn.score(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.predict([[22]])
```

```python
lr_predict=lr.predict([[22]])
```

```python
print('For LinearRegression accuracy : ',lr_predict)
print('For KNN accuracy : ',knn_predict)
```

```python
lr.score(x,y)
```

```python
data['predict'] = lr.predict(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,data.predict)
plt.scatter(x,y)
```
