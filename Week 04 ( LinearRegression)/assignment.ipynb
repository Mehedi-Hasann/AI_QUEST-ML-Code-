```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vegetable_data = {
    'weight' : [2,4,5,3,6,5,7],
    'price' : [35,60,20,50,50,55,60]
}
```

```python
df = pd.DataFrame(vegetable_data)
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
x = df[['weight']]
```

```python
y = df[['price']]
```

```python
x
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[6]])
```

```python
df['predict'] = reg.predict(x)
```

```python
df
```

```python
df['Residuals'] = df.price - df.predict
```

```python
df
```

```python
plt.scatter(df.weight,df.price,marker='o',color='blue')
plt.plot(df.weight,df.predict)
plt.xlabel('Weight')
plt.ylabel('Price')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(df.predict,df.price)
```
