Linear Regression using Gradient Descent

```python
m = 0
c = 0
lr = 0.0001
n = len(x)

for i in range(500):
    yhat = m*x + c
    Dm = (-2/n)*sum(x*(y-yhat))
    Dc = (-2/n)*sum(y-yhat)

    m = m-(lr*Dm)
    c = c -(lr*Dc)
    print(yhat)
```

```python
print(m,c)
```
